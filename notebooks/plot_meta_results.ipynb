```python
import pickle
import matplotlib.pyplot as plt

from configs import BaseConfig
config = BaseConfig()
```

```python
def load_data(results_path):
    results = []
    for result_file in results_path.glob('*.pkl'):
        with open(result_file, 'rb') as f:
            result = pickle.load(f)
        results.append(result)
    return results

def plot_results(results):
    fig, axs = plt.subplots(1, 2, sharex=True, figsize=(10, 4))
    for r in results:
        axs[0].plot(r['rwd'], label=round(r['noise'][0], 2))
        axs[1].plot(r['mae'], label=round(r['noise'][0], 2))
    axs[0].set_ylabel('Reward')
    axs[1].set_ylabel('MAE')
    plt.legend()
    plt.tight_layout()
    plt.show()
```

```python
results_path = config.results_path / 'noise_only'
results = load_data(results_path)
plot_results(results)
```
